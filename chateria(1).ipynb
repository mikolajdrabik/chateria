Zadanie 1.
Które z następujących zadań wymagają w Twojej opinii inteligencji od człowieka;


* Odpowiadanie na proste pytania zadawane w języku naturalnym (np. polskim) - Sztuczna inteligencja nie potrafi odbierać oraz interpretować czyichś zamiarów oraz emocji, jakie ktoś wyraża podczas zadawania pytania, a to może zmienić kontekst wypowiedzi.
* komponowanie muzyki (AI nie potrafi komponować muzyki w która w naszych kryteriach nie jest uznawana za dobrą czy też ambitną, nie potrafi również wyrażać przez nią żadnych emocji które są niezbędnyn elementem w wszelkiej maści utworach a to czyni ją bezpłciową.

Zadanie 2. Które z następujących problemów można uznać za mieszczące się w zakresie sztucznej
inteligencji

* streszczanie tekstu (AI robi to znacznie szybciej od człowieka i może pracować na ogromnych bazach danych)
* rozwiązywanie układów równań (potrafi wykonać znacznie bardziej skomplikowane działania niż człwoiek)  
* rozwiązywanie układów równań liniowych (potrafi wykonaż znacznie bardziej skomplikowane działania niż człwoiek oraz robi to znacznie sprawniej (posiada większą moc obliczeniową))
* symboliczne obliczanie pochodnych (potrafi wykonać znacznie bardziej skomplikowane działania od człowieka)
* symboliczne całkowanie (potrafi wykonać znacznie bardziej skomplikowane działania niż człwoiek)
* kierowanie samochodem (AI jest w stanie prowadzić samochód ale z racji tego że inni kierowcy sż czasami nieprzewidywalni maszyna nie jest w stanie przewidywać niektorych zdarzeń na które człowiek jest przystosowany)


Zadnie 3. Które z poniższych rodzajów komunikacyjnego zachowania człowieka mogą być
obecnie skutecznie imitowane przez sztuczne systemy (odpowiednio oprogramowane
maszyny):




* dyskusja polityczna
* dyskusja naukowa  
Ai potrafi skutecznie wykrywać błedy logiczne jakie ktoć wprowadza do dsykusji oraz wyróznia sie mysleniem dedukcyjnym przez co jest w stanie skutecznie odpierac argumenty z innej ze stron i skutecznie usasadniać swoje.

Zadanie 4.
Do zadań skorzystaj z załączonej listy Chatbotów (Chatterbotów) lub znajdź własne.

1. Przeprowadź rozmowę z chatbotem. Spróbuj zdefiniować różnice pomiędzy
botem udającym człowieka (przygotowywanym na test Turinga) a botem
„asystentem, służącym”:



*   Bot przygotowany do testu Turinga bedzie odpowiadał w sposób bardziej naturalny oraz zakres pytań jakie możemy mu zadać będzie znacznie większy niż u botów asystentów.
*   Bot asystent głównie skupia się na pomocy użytkownikom którzy korzystają z pomocy bota. Tego rodzaju bot bedzie wypowiadał się w sposób mało naturalny a odpowiedzi nie związane z tematem do którego jest przeznaczony bedzie stanowczo odrzucał.

2. Sprawdź dwa boty z obu z tych rodzajów na występowanie zachowań:

a) opowiadanie żartów,

Cleverbot bedzię starał opowiadać żart najczęsciej o prostej koncepcji.

Ania z Ikea stanowczo odmówi prośby dotyczącej opwiedzenie żartu.

b) przytaczanie cytatów z twoich wypowiedzi, lub znanych osób,

Cleverbot potrafi cytować, czesto są to losowe rzeczy.

Ania z Ikea odmówi zacytowanie czego kolwiek.


c) nawiązywanie wypowiedzi do słów kluczowych,

Cleverbot często reaguje na słwoa kluczowe lecz czasami odpowiedzi wydają sie niepowiązane

Anie z Ikea bedzie prawidłowo reagować na słowa kluczowe


d) zadawanie dużej liczby pytań,

Cleverbot czesto bedzie zadawac pytania po to by zachować ciągłość rozmowy

Ania z Ikea zada pytania jedynie wtedy, gdy będzie to konieczne do realizacji zadania

e) powracanie do początku wypowiedzi, sekwencyjne powtarzanie,

Cleverbot potrafi wracać do wcześniejszych sekwencji oraz wątków

f) zadawanie pytań powstających z twoich wypowiedzi,

Cleverbot potrafi generować pytania z moich wypowiedzi lecz mogą wydawać sie one nieco losowe

Ania z Ikea zapyta sie tylko wtedy kiedy bedzie to konieczne do rozwiązania jakiegoś problemu

g) odpowiadanie wymijająco, ogólnikowo,

Cleverbot robi to bardzo często

Ania z Ikea bedzie odpowiadać stosunkowo precyzyjnie

h) częsta zmiana tematu rozmowy,

Cleverbot robi to bardzo często

Ania z Ikea bedzie starała się pozostać przy jednym temacie

i) problemy z utrzymaniem wątków

Cleverbot ma problemy z utrzymanie spójności rozmowy

Ania z Ikea podtrzymuje stale jeden wątek
