Zadanie 1.
Które z następujących zadań wymagają w Twojej opinii inteligencji od człowieka;


* Odpowiadanie na proste pytania zadawane w języku naturalnym (np. polskim) - Sztuczna inteligencja jest w stanie odpowiadać na protste pytania lecz nie potrafi odbierać oraz interpretować czyichś zamiarów oraz emocji, jakie ktoś wyraża podczas zadawania pytania tak jak inteprtuje to człowiek, a to może zmienić kontekst wypowiedzi.
* komponowanie muzyki - AI nie potrafi komponować muzyki która w naszych kryteriach nie jest uznawana za dobrą czy też ambitną, nie potrafi również wyrażać przez nią żadnych emocji które są niezbędnyn elementem w wszelkiej maści utworach przez co czyni ją bezpłciową.

inteligencji

* rozwiązywanie układów równań
* rozwiązywanie układów równań liniowych
* symboliczne obliczanie pochodnych
* symboliczne całkowanie
(potrafi wykonaż znacznie bardziej skomplikowane działania niż człwoiek oraz robi to znacznie sprawniej, posiada również większą moc obliczeniową a to wszystko może się przyczynić do szybkiego rozwoju w dziedzinie matematyki)

* kierowanie samochodem (AI jest w stanie prowadzić samochód ale z racji tego że inni kierowcy są czasami nieprzewidywalni maszyna nie jest w stanie kalkulować niektorych zdarzeń do których człowiek jest przyzwyczajony co może satonwić dalszy przedmiot badań który bedzie miał na celu utworzenie sztucznej inteligencji jako postać idealnego kierowcy)


Zadnie 3. Które z poniższych rodzajów komunikacyjnego zachowania człowieka mogą być
obecnie skutecznie imitowane przez sztuczne systemy (odpowiednio oprogramowane
maszyny):

Rozmowa towarzyska - Sztuczne systemy potrafią z łatwością prowadzić proste rozmowy towarzyskie, jednak mają określone kryteria na jakie tamaty mogą rozmowiać. Przykładem są chatboty i asystenci głosowi.


Dyskusja naukowa i dyskusja polityczna - Ai potrafi skutecznie wykrywać błedy logiczne jakie ktoś wprowadza do dsykusji oraz wyróznia sie mysleniem dedukcyjnym przez co jest w stanie skutecznie odpierać argumenty z innej ze stron i skutecznie usasadniać swoje jednak prowadzenie twórczej dyskusji naukowej wymaga głębszego rozumienia kontekstu, posiadania swojego zdania oraz krytycznego myślenia, co nadal jest wyzwaniem.

Odpowiadanie na pytania klientów w telefonicznej infolinii - Tak. Systemy IVR (Interactive Voice Response) oraz asystenci głosowi coraz skuteczniej odpowiadają na typowe pytania klientów, pomagając w rozwiązywaniu problemów.

Odpowiadanie na pytania klientów w internetowej infolinii - Boty występują na większości infoliń, są powszechnie stosowane i często skutecznie radzą sobie z prostymi pytaniami klientów.

* dyskusja polityczna
* dyskusja naukowa  

Ai potrafi skutecznie wykrywać błedy logiczne jakie ktoś wprowadza do dsykusji oraz wyróznia sie mysleniem dedukcyjnym przez co jest w stanie skutecznie odpierać argumenty z innej ze stron i skutecznie usasadniać swoje.

Zadanie 4.
Do zadań skorzystaj z załączonej listy Chatbotów (Chatterbotów) lub znajdź własne.

1. Przeprowadź rozmowę z chatbotem. Spróbuj zdefiniować różnice pomiędzy
botem udającym człowieka (przygotowywanym na test Turinga) a botem
„asystentem, służącym”:



*   Bot przygotowany do testu Turinga bedzie odpowiadał w sposób bardziej naturalny oraz zakres pytań jakie możemy mu zadać będzie znacznie większy niż u botów asystentów.
*   Bot asystent głównie skupia się na pomocy użytkownikom którzy korzystają z pomocy bota. Tego rodzaju bot bedzie wypowiadał się w sposób mniej naturalny niż boty przygotowane do testu turinga a pytania niezwiązane z tematem do którego jest przeznaczony bedzie stanowczo odrzucał.

2. Sprawdź dwa boty z obu z tych rodzajów na występowanie zachowań:

a) opowiadanie żartów,

Cleverbot: Bot opowie żart jeśli go o to  poprosimy i nie bedzie mial z tym większego problemu.

Billie z Ikea: Stanowczo odmówi opowiedzenia żartu.

b) przytaczanie cytatów z twoich wypowiedzi, lub znanych osób,

Cleverbot: Bot bez problemu cytuję wypowiedzi moje, znanych osób oraz różne kwestie z filmów czy ksiażek.

Billie z Ikea: Bot odmawia zacytowania czegokoliwek ponieważ uaważa to za nieistotny wątek


c) nawiązywanie wypowiedzi do słów kluczowych,

Cleverbot: Potrafi nawiązywać do słów kluczowych na przykład w celu podtrzymania tematu rozmowy.

Billie z Ikea: Potrafi nawiązywać do słów kluczowych ale tylko w kontekście pomocy.

d) zadawanie dużej liczby pytań,

Cleverbot: Bardzo często zadaję pytanie lecz niektóre z nich są bardzo często losowe.

Billie z Ikea: Zada pytanie tylko jeśli jest to koniczne do rozwiązania problemu lub w celu potwierdzenia czegoś.

e) powracanie do początku wypowiedzi, sekwencyjne powtarzanie,

Cleverbot potrafi wracać do wcześniejszych sekwencji oraz wątków

Billie z Ikea: Robi to rzadko albo wcale jeżeli nie jest to koniczne

f) zadawanie pytań powstających z twoich wypowiedzi,

Cleverbot: Potrafi generować pytania z moich wypowiedzi lecz mogą wydawać sie one nieco losowe.
Billie z Ikea: Zapyta sie tylko wtedy kiedy bedzie to konieczne do rozwiązania jakiegoś problemu.

g) odpowiadanie wymijająco, ogólnikowo,

Cleverbot: Robi to bardzo często.

Billie z Ikea: Bedzie odpowiadać stosunkowo precyzyjnie jeśli pytanie jest zgodne z pewnym kryterium.

h) częsta zmiana tematu rozmowy,

Cleverbot: Często zadaje losowe pytanie przez co zmienia tamat rozmowy.

Billie z Ikea: Bedzie bardzo restrykcyjnie trzymał się tamatu o który go zapytaliśmy lecz tamat musi być związany z pomocą jaką chcemy uzyskać.

i) problemy z utrzymaniem wątków

Cleverbot: Ma problemy z utrzymanie spójności rozmowy, często zmienia temat zadająć pytania niezwiązane z danym wątkiem lub odpowiada ogólnikami.

Billie z Ikea: Podtrzymuje stale jeden wątek który związany jest z udzielaniem pomocy.
